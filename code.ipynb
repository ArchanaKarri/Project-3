```python
# Dependencies and Setup
import pandas as pd
import pandas as geek

# File to Load (Remember to Change These)
production_data_to_load = "Resources/shipment_data_2021-2022_financial_year.csv"


# Production data reading  and store into Pandas DataFrames
production_data_df = pd.read_csv(production_data_to_load)


```

```python
#Displaying the DataFrame
production_data_df.head()
```

```python
# Data types od all columns in the DataFrame
production_data_df.dtypes
```

```python
#dropping rows with null values in smpPlantID,smlPartWarehouseLocationID, impPartGroupID
prod_data_df = production_data_df.dropna(subset=['smpPlantID','smlPartWarehouseLocationID', 'impPartGroupID'])
```

```python
prod_data_df
```

```python
# Adding SmlQuantityShipped and smlJobQuantityShipped into another column
prod_data_df['quantityShipped']=prod_data_df['smlQuantityShipped']+prod_data_df['smlJobQuantityShipped']
```

```python
prod_data_df
```

```python
#Dropping the columns smlQuantityShipped , smlJobQuantityShipped as their sum is already stored
prod_data_df_new=prod_data_df.drop(columns=['smlQuantityShipped','smlJobQuantityShipped'])
```

```python
prod_data_df_new
```

```python

df_new = prod_data_df_new.drop(prod_data_df_new[(prod_data_df_new['smpReversed']==True) | (prod_data_df_new['smpReversalEntry']==True)].index)
```

```python
df_new
```

```python
# Removing columns smlPartWarehouseLocationID, smpReversed,smpReversalEntry,smpShipLocationID from the dataframe
df_new=df_new.drop(columns=['smlPartWarehouseLocationID','smpReversed','smpReversalEntry','smpShipLocationID'])
```

```python
df_new.dtypes
```

```python
# Converting 'impPartGroupID' data type to String
df_new['impPartGroupID']  = df_new['impPartGroupID'] .astype("string")
#Filtering data from 'impPartGroupID' with few selected values into df_new_clean dataframe
df_new_clean = df_new.loc[(df_new['impPartGroupID']=="BED") | (df_new['impPartGroupID'] =="BINS") |(df_new['impPartGroupID'] =="BULK") |(df_new['impPartGroupID'] =="CARD") |(df_new['impPartGroupID'] =="CASE") |(df_new['impPartGroupID'] =="PAL") |(df_new['impPartGroupID'] =="PAL2") |(df_new['impPartGroupID'] =="PEN") ]
```

```python
df_new_clean
```

```python
# Writing the result into Clean.csv file
df_new_clean.to_csv('clean.csv')
```
